```python
import random
```

```python
class TicTacToe():
    """Tic Tac Toe Game"""
    
    def __init__ (self):
        """Initialize Tic Tac Toe Board"""
        self.board = []
        self.winner = '_'
        for i in range(9):
            self.board.append('_')
            

    def printBoard (self):
        print ("BOARD")
        print ("=====")
        for i in range(9):
            print (self.board[i] + ' ', end="") 
            if (i%3 == 2):
              print ('')

    def printPossibleNextMove (self):
        print ("Possible Next Move")
        print ("==================")
        for i in range(9):
            if (self.board[i] == '_'):
                print (str(i) + ' ', end="")
            else:
                print (' ' + ' ', end="")
            if (i%3 == 2):
              print ('')
              
            
    def placeMove (self, player, move):
        if (move in range(9) and len (player) == 1 and (player.lower()=='o' or player.lower() == 'x')):
            if (self.board[move] == '_'):   #making sure it is emptied.
                self.board[move] = player

    def undoMove (self, player, move):
        if (self.board[move] == player):
            self.board[move] = '_'
        self.isGameOver()

    def otherPlayer (self, player):
        if (player.lower()=='o'):
            return 'X'
        else:
            return 'O'


    def test (self):
        
        print ("BEGIN - TEST METHOD otherPlayer")
        print ("Other Player of o is " + self.otherPlayer('o'))
        print ("Other Player of O is " + self.otherPlayer('O'))
        print ("Other Player of x is " + self.otherPlayer('x'))
        print ("Other Player of X is " + self.otherPlayer('X'))
        print ("END - TEST METHOD otherPlayer")
        
        print ("BEGIN - TEST METHOD undoMove")
        self.printBoard()
        self.placeMove ('X', 5)
        self.printBoard()
        print ("  UndoMove 5 ")
        self.undoMove ('X', 5)
        self.printBoard()

        self.printBoard()
        self.placeMove ('X', 5)
        self.printBoard()
        print ("  UndoMove X, 4 - negative test case")
        self.undoMove ('X', 4)

        print ("  UndoMove O, 5 - negative test case")
        self.undoMove ('O', 5)
        self.printBoard()
        self.undoMove ('X', 5)
        self.printBoard()
        print ("END - TEST METHOD undoMove")

        print ("BEGIN - TEST METHOD simulateBestMove")
        self.printBoard()
        self.placeMove ('X', 0)
        self.placeMove ('X', 1)
        self.placeMove ('O', 3)
        self.placeMove ('O', 4)
        self.placeMove ('X', 6)        
        self.placeMove ('O', 7)
        self.printBoard()

#        print ("Simulate X on 2: " + str(self.simulateBestMove ('X', 2)))
#        self.printBoard()
        
        print ("Simulate O on 2: " + str(self.simulateBestMove ('O', 2)))
        self.printBoard()

        print ("Simulate O on 8: " + str(self.simulateBestMove ('O', 8)))
        self.printBoard()

#        print ("Simulate X on 5: " + str(self.simulateBestMove ('X', 5)))
#        self.printBoard()

        print ("END - TEST METHOD simulateBestmove")
        
                 
    #
    # Return -1 if loss, 0 if tie, 1 if winning
    # 
    def simulateBestMove (self, player, move):
        result = -1

        #self.printBoard()
        #print ("simulating " + player + " " + str(move))
        self.placeMove (player, move)
        #self.printBoard()
        
        if (self.isGameOver()):
            if (self.winner == '_'):
                result = 0
            elif (self.winner == player):
                result = 1
            else:
                result = -1
        else:
            # Explore all the other moves for the opponent.
            # Save only the best move for the opponent
            #
            bestOpponentResult = -1
            for i in range(9):
                if (self.board[i] == '_'):
                    opponentResult = self.simulateBestMove (self.otherPlayer(player), i)
                    if (opponentResult > bestOpponentResult):
                        bestOpponentResult = opponentResult
                    if (bestOpponentResult == 1):
                        break
                            
            if (bestOpponentResult == 1):
                result = -1
            elif (bestOpponentResult == -1):
                result = 1
            else:
                result = 0
            # print ("Best Opponent Result " + self.otherPlayer(player) + " " + str (bestOpponentResult))
            
        self.undoMove (player, move)
        
        # print ("simulating - undo " + player + " " + str(move))
        # self.printBoard()
        
        return result
        
                
    def getBestMove (self, player):
        winningMove = -1
        result = -1        
        for i in range(9):
            if (self.board[i] == '_'):
                currentResult = self.simulateBestMove (player, i)
                print ("Checking step: " + str(i) + ", Result = " + str(currentResult))
                if (currentResult == 1 or currentResult == 0):
                    result = currentResult
                    winningMove = i
                if (currentResult == 1):
                    break

        print ("Best Move is : " + str(winningMove))
        self.placeMove (player, winningMove)                

                 
    def getUserMove (self):
         done = False
         while (not done):
             move = input ("Player X, please enter your next move: ")
             move = int(move)
             if (move in range(9) and self.board[move] == '_'):
                 self.placeMove ('X', move)
                 done = True

    # Get random computer move                
    def getComputerMove (self):
         done = False
         while (not done):
             move = random.randint(0,8)
             if (move in range(9) and self.board[move] == '_'):
                 self.placeMove ('O', move)
                 print ("Computer placing move on: " + str(move))
                 done = True

        

    def isGameOver (self):
        self.winner = '_'
        winner = '_'
        gameOver = False


        # check the rows
        row = 0
        for i in range(3):
            if (self.board[row] != '_' and
                self.board[row] == self.board[row+1] and
                self.board[row] == self.board[row+2]):
                
                gameOver = True
                self.winner = self.board[row]
                return gameOver
            row += 3

        # check the columns
        col = 0
        for i in range(3):
            if (self.board[col] != '_' and
                self.board[col] == self.board[col+3] and
                self.board[col] == self.board[col+6]):
                
                gameOver = True
                self.winner = self.board[col]
                return gameOver
            col += 1

        # check \
        if (self.board[0] != '_' and
            self.board[0] == self.board[4] and
            self.board[0] == self.board[8]):
                gameOver = True
                self.winner = self.board[4]
                return gameOver

        # check /            
        if (self.board[2] != '_' and
            self.board[2] == self.board[4] and
            self.board[2] == self.board[6]):
                gameOver = True
                self.winner = self.board[4]                
                return gameOver

        # check any position left
        for i in range(9):
            if (self.board[i] == '_'):  #found open position
               break
            if (i == 8):                #no open position found
               gameOver = True

        return gameOver
             
                 
    def startGame (self):
        print ("Tic Tac Toe Game")
        print ("Computer = 'O', User = 'X'")
        userMove = True

        while (not self.isGameOver()):
            self.printBoard()
            self.printPossibleNextMove()
            if (userMove):
              self.getUserMove()
              userMove = False
            else:
              # self.getComputerMove()
              self.getBestMove('O')
              userMove = True

        self.printBoard()
        
        if (self.winner == '_'):
            print ("The game is tie.")
        elif (self.winner == 'O'):
            print ("Sorry, you lost. Perhaps you have better luck next time")
        else:
            print ("Congratulation, you have won. The winner is " + self.winner)
        
   
    
myTTT = TicTacToe()
# myTTT.test()
myTTT.startGame()
```
